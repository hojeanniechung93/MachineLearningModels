# Using Images To Explore Clustering

```python
# This notebook illustrates some examples of using k-means, DBSCAN, and
# agglomerative clustering to cluster color in an image. After running this
# introductory bit of code, each section can be run independently. Because
# of this, there is code that gets repeated in each section. Parts of the
# code that are noteworthy are commented; however, in following sections
# only new noteworthy comments are made.

# Allow us to see plots in the notebook
%matplotlib inline

# Libraries to read and plot images
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread('C:/Temp/ColorfulDuck.jpg')
```

```python
plt.imshow(img)
```

```python
img.shape
```

## Partition Clustering (K-Means)
### Six cluster example

```python
# Turn our image into a feature array for clustering - one pixel per row
nRows = img.shape[0]
nCols = img.shape[1]
imgBGR = img.flatten().reshape(nRows * nCols, 3)
```

```python
# Do the k-means clustering
from sklearn.cluster import KMeans

nClusters = 6
kmeans = KMeans(n_clusters = nClusters, random_state = 0).fit(imgBGR)
```

```python
# Get the cluster centers and labels
clusterCenters = np.around(kmeans.cluster_centers_).astype(np.uint8)
clusterLabels = kmeans.labels_
```

```python
# Replace original color in the image with the representative color for
# the cluster
for i in range(nClusters):
    imgBGR[clusterLabels == i] = clusterCenters[i]
```

```python
# Put the data back into image form
imgKmeans = imgBGR.flatten().reshape(nRows, nCols, 3)
```

```python
# See our result
plt.imshow(imgKmeans)
```

### Increasing the number of clusters

```python
# Go through the process again but with requesting more clusters
nRows = img.shape[0]
nCols = img.shape[1]
imgBGR = img.flatten().reshape(nRows * nCols, 3)
```

```python
from sklearn.cluster import KMeans

nClusters = 30
kmeans = KMeans(n_clusters = nClusters, random_state = 0).fit(imgBGR)
```

```python
clusterCenters = np.around(kmeans.cluster_centers_).astype(np.uint8)
clusterLabels = kmeans.labels_
```

```python
for i in range(nClusters):
    imgBGR[clusterLabels == i] = clusterCenters[i]
```

```python
imgKmeans = imgBGR.flatten().reshape(nRows, nCols, 3)
```

```python
plt.imshow(imgKmeans)
```

## Density Clustering (DBSCAN)
### Clustering just the color space (similar to the k-means example)

```python
nRows = img.shape[0]
nCols = img.shape[1]
imgBGR = img.flatten().reshape(nRows * nCols, 3)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=2, min_samples=30).fit(imgBGR)
```

```python
clusterLabels = dbscan.labels_
nClusters = np.unique(clusterLabels).size
```

```python
# Since we don't have cluster centers from DBSCAN, use the average color as
# the cluster center. Also, DBSCAN does not cluster everything, so mark those
# as magenta so we can see what did  not get clustered
clusterCenters = np.zeros((nClusters, 3))
for i in range(nClusters - 1):
    clusterCenters[i] = np.mean(imgBGR[clusterLabels == i], axis = 0)

clusterCenters[nClusters - 1] = (255, 0, 255)

clusterCenters = np.around(clusterCenters).astype(np.uint8)
```

```python
for i in range(nClusters - 1):
    imgBGR[clusterLabels == i] = clusterCenters[i]
    
imgBGR[clusterLabels == -1] = clusterCenters[nClusters - 1]
```

```python
imgDBSCAN = imgBGR.flatten().reshape(nRows, nCols, 3)
```

```python
plt.imshow(imgDBSCAN)
```

### Adding pixel coordinates as features

```python
# This time we are going to incorporate pixel coordinates, so leave room for
# those
nRows = img.shape[0]
nCols = img.shape[1]
imgBGR = np.zeros((nRows * nCols, 5))

# Stretch the pixel coordinates some to allow room for grouping a larger
# range of colors
imgBGR[:, 2:5] = img.flatten().reshape(nRows * nCols, 3)
imgBGR[:, 0] = np.array([[i] * nCols for i in range(nRows)]).flatten() * 2
imgBGR[:, 1] = np.array([i for i in range(nCols)] * nRows) * 2
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=12, min_samples=30).fit(imgBGR)
```

```python
clusterLabels = dbscan.labels_
nClusters = np.unique(clusterLabels).size
```

```python
# Since we don't have cluster centers from DBSCAN, use the average color as
# the cluster center. Also, DBSCAN does not cluster everything, so mark those
# as magenta so we can see what did not get clustered
clusterCenters = np.zeros((nClusters, 3))
for i in range(nClusters - 1):
    clusterCenters[i] = np.mean(imgBGR[clusterLabels == i, 2:5], axis = 0)

clusterCenters[nClusters - 1] = (255, 0, 255)

clusterCenters = np.around(clusterCenters).astype(np.uint8)
```

```python
imgBGR = np.zeros((nRows * nCols, 3), dtype = np.uint8)
for i in range(nClusters - 1):
    imgBGR[clusterLabels == i] = clusterCenters[i]
    
imgBGR[clusterLabels == -1] = clusterCenters[nClusters - 1]
```

```python
imgDBSCAN = imgBGR.flatten().reshape(nRows, nCols, 3)
```

```python
plt.imshow(imgDBSCAN)
```

## Hierarchical Clustering (Agglomerative)
### Clustering just the color space

```python
import scipy as sp

# Vanilla agglomerative clustering does not deal well with large datasets
# (it needs to calculate a total pixels by total pixels similarity matrix),
# so start with a small image.
imgSmall = sp.misc.imresize(img, 0.5)
nRows = imgSmall.shape[0]
nCols = imgSmall.shape[1]

imgBGR = imgSmall.flatten().reshape(nRows * nCols, 3)
```

```python
from sklearn.cluster import AgglomerativeClustering

# We don't have to split the tree at a fixed merge distance and can instead
# choose the number of clusters giving us something more similar to what
# k-means did
nClusters = 6
agg = AgglomerativeClustering(n_clusters = nClusters, linkage = 'ward').fit(imgBGR)
```

```python
clusterLabels = agg.labels_
```

```python
clusterCenters = np.zeros((nClusters, 3))
for i in range(nClusters):
    clusterCenters[i] = np.mean(imgBGR[clusterLabels == i], axis = 0)
clusterCenters = np.around(clusterCenters).astype(np.uint8)
```

```python
for i in range(nClusters):
    imgBGR[clusterLabels == i] = clusterCenters[i]
```

```python
imgAgg = imgBGR.flatten().reshape(nRows, nCols, 3)
```

```python
plt.imshow(imgAgg)
```

### Leveraging the image graph (relationship between pixels)

```python
import scipy as sp

nRows = img.shape[0]
nCols = img.shape[1]

imgBGR = img.flatten().reshape(nRows * nCols, 3)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction.image import grid_to_graph

# Even more powerful is to leverage the connectivity graph of the pixels.
# We could introduce pixel coordinates similar to in DBSCAN, but leveraging
# the inherent structure in the data also relaxes the need for the similarity
# matrix (note we did not scale the image down in this case)
connectivity = grid_to_graph(*img.shape[0:2])

# Since similar colors separated by distance will not be the same cluster,
# we will need more clusters. One could go en extra step after this and
# cluster the cluster colors (with something like k-means) to reduce the
# number of colors.
nClusters = 30
agg = AgglomerativeClustering(n_clusters = nClusters, linkage = 'ward',
                             connectivity = connectivity).fit(imgBGR)
```

```python
clusterLabels = agg.labels_
```

```python
clusterCenters = np.zeros((nClusters, 3))
for i in range(nClusters):
    clusterCenters[i] = np.mean(imgBGR[clusterLabels == i], axis = 0)
clusterCenters = np.around(clusterCenters).astype(np.uint8)
```

```python
for i in range(nClusters):
    imgBGR[clusterLabels == i] = clusterCenters[i]
```

```python
imgAgg = imgBGR.flatten().reshape(nRows, nCols, 3)
```

```python
plt.imshow(imgAgg)
```
