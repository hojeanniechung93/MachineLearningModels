```python
## Decision trees are the building blocks of some of the most powerful supervised learning methods that are used today. 
A decision tree is basically a binary tree flowchart where each node splits a group of observations according to some feature variable. 
The goal of a decision tree is to split your data into groups such that every element in one group belongs to the same category. 

We will show you how to generate a decision tree and create a graph of it in a Jupyter Notebook.
In this example, we will be using the classic iris dataset. Use the following code to load it:
```

```python
import sklearn.datasets as datasets
import pandas as pd
import sklearn.cross_validation as cv
import numpy as np

iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
```

```python
df = pd.DataFrame([np.append(e, f) for e, f in zip(iris.data, iris.target)], columns=np.append(iris.feature_names, 'target'))
df.head()
```

```python
X_train, X_test, y_train, y_test = cv.train_test_split(df, y, test_size=0.33, random_state=42)
y_train
```

### Below is the decision trees estimation

```python
df.as_matrix()
```

```python
y
```

```python
Sklearn will generate a decision tree for your dataset using an optimized version of the CART algorithm when you run the following code.
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
You can also import DecisionTreeRegressor from sklearn.tree if you want to use a decision tree to predict a numerical target variable.
Try switching one of the columns of df with our y variable from above and fitting a regression tree on it.
Now that we have a decision tree, we can use the pydotplus package to create a visualization for it.
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
import sklearn.metrics as met

y_pred = dtree.predict(X_test)

print(met.classification_report(y_test, y_pred))
```
